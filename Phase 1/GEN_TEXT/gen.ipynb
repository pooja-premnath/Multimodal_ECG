```python
import pandas as pd
import ast
df=pd.read_csv("ptbxl_database.csv")
df.head()
```

```python

df.scp_codes = df.scp_codes.apply(lambda x: ast.literal_eval(x))
```

```python
agg_df=pd.read_csv("scp_statements.csv",index_col=0)
agg_df=agg_df[agg_df.diagnostic == 1]
```

```python
def aggregate_diagnostic(y_dic):
    tmp = []
    for key in y_dic.keys():
        if key in agg_df.index:
            tmp.append(agg_df.loc[key].diagnostic_class)
    return list(set(tmp))
```

```python
df['diagnostic_superclass'] = df.scp_codes.apply(aggregate_diagnostic)
```

```python
def generate_prompt(row):
    diagnostic_superclass_mapping = {
    '[NORM]': 'Normal ECG',
    '[MI]': 'Myocardial Infarction',
    '[STTC]': 'ST/T Change',
    '[CD]': 'Conduction Disturbance',
    '[HYP]': 'Hypertrophy'
}
    # sex = 'male' if row['sex'] == 1 else 'female'
    report = row['report']
    heart_axis = row['heart_axis'] if not pd.isnull(row['heart_axis']) else 'no heart axis data'
    infarction = "No infarction detected" if pd.isnull(row['infarction_stadium1']) else "Infarction detected"
    diagnostic_superclass_raw = row['diagnostic_superclass']
    
    # Constructing the prompt with diagnostic_superclass as-is
    prompt = (f"Generate a concise 1-line ECG report for a patient. "
              f"The report mentions: '{report}', heart axis as {heart_axis}, "
              f"diagnostic superclass: {diagnostic_superclass_raw}, and {infarction}. Generate only one line report nothing else. ")
    
    return prompt
```

```python
# run for row 1
row = df.iloc[7]
print(row["heart_axis"])
print(generate_prompt(row))
```

```python
from langchain_community.llms import Ollama

llm = Ollama(model="llama3.1")

def get_model_response(user_prompt, system_prompt, temperature=0.2):  # Default to lower temperature for consistency
    prompt = f"""
        <|begin_of_text|>
        <|start_header_id|>system<|end_header_id|>
        { system_prompt }
        <|eot_id|>
        <|start_header_id|>user<|end_header_id|>
        { user_prompt }
        <|eot_id|>
        <|start_header_id|>assistant<|end_header_id|>
        Please ensure that all responses follow this structure:
        - Provide a brief introduction, outlining the main topic.
        - Present key details in a structured, bullet-point format if applicable.
        - End with a concise conclusion or next steps, depending on the query.
        <|eot_id|>
    """
    response = llm.invoke(prompt, temperature=temperature)
    return response
```

```python

def generate_report(row):
    user_prompt = generate_prompt(row)
    print(user_prompt)
    system_prompt = "Strictly 1 line only. Generate only report nothing else. Start main report with | and end with |. Do not include any other information. Do not use the pghrase 'ECG report' in the report."
    response = get_model_response(user_prompt, system_prompt)
    return response
```

```python

row = df.iloc[10]
print(generate_report(row))
```

```python
import pandas as pd

# Load the two files: your main DataFrame (ptbxl_database) and the abbreviation descriptions (scp_statements.csv)
df_scp_statements = pd.read_csv('scp_statements.csv')  # Replace with your actual file path
ptbxl_df = pd.read_csv('ptbxl_database.csv')  # Replace with your actual file path
import pandas as pd
import ast  # For safely evaluating string representations of dictionaries

# Load the two files: your main DataFrame (ptbxl_database) and the abbreviation descriptions (scp_statements.csv)
# df_scp_statements = pd.read_csv('path_to_your_scp_statements.csv')  # Replace with your actual file path
# ptbxl_df = pd.read_csv('path_to_your_ptbxl_database.csv')  # Replace with your actual file path

# Rename 'Unnamed: 0' to 'abbr' for abbreviation
df_scp_statements = df_scp_statements.rename(columns={'Unnamed: 0': 'abbr'})

# Create a dictionary mapping abbreviations to descriptions
description_dict = df_scp_statements.set_index('abbr')['description'].to_dict()

# Function to concatenate descriptions for abbreviations with values > 0
def concatenate_descriptions_above_zero(code_str):
    try:
        # Safely evaluate the string into a dictionary
        code_dict = ast.literal_eval(code_str)
    except (ValueError, SyntaxError):
        # If the string cannot be parsed, return an empty string or the original string
        return ''
    
    # Now check for keys with values > 0
    keys_above_zero = [key for key, value in code_dict.items() if value > 0]
    descriptions = [description_dict.get(key, key) for key in keys_above_zero]  # Get description or fallback to key if missing
    return ' '.join(descriptions)

# Apply this function to the 'scp_codes' column of your dataframe and create a new column 'report_with_descriptions'
ptbxl_df['report_with_descriptions'] = ptbxl_df['scp_codes'].apply(concatenate_descriptions_above_zero)

# View the updated dataframe
ptbxl_df[["ecg_id",'scp_codes', 'report_with_descriptions'.'filename_lr','filename_hr']].to_csv("ptbxl_database_with_descriptions.csv")
```

```python
import os

def rename_png_files_in_subfolders(base_path):
    try:
        # Iterate over each subfolder inside the base path
        for folder_name in os.listdir(base_path):
            folder_path = os.path.join(base_path, folder_name)

            if os.path.isdir(folder_path):  # Ensure it's a directory
                print(f"Processing folder: {folder_path}")

                # Iterate over the files in each folder
                for file_name in os.listdir(folder_path):
                    old_file_path = os.path.join(folder_path, file_name)

                    # Check if it's a file and a PNG file
                    if os.path.isfile(old_file_path) and file_name.endswith(".png"):
                        # Check if "-0" is in the filename before renaming
                        if "-0" in file_name:
                            new_file_name = file_name.replace("-0", "")
                            new_file_path = os.path.join(folder_path, new_file_name)

                            # Rename the file
                            os.rename(old_file_path, new_file_path)
                            print(f"Renamed file: {old_file_path} -> {new_file_path}")
                        else:
                            print(f"No renaming needed for: {file_name}")
                    else:
                        # Delete non-PNG files
                        if os.path.isfile(old_file_path):
                            os.remove(old_file_path)
                            print(f"Deleted non-PNG file: {old_file_path}")
    except PermissionError as e:
        print(f"PermissionError: {e}. Could not access folder or files.")
    except Exception as e:
        print(f"Error: {e}")

# Set the base path to your folder containing the subdirectories


####### CHANGE PATH #########
base_path = "/content/drive/MyDrive/ptbxl/generated_ecg_images"
rename_png_files_in_subfolders(base_path)
```

```python
class AvgMeter:
    def __init__(self, name="Metric"):
        self.name = name
        self.reset()

    def reset(self):
        self.avg, self.sum, self.count = [0] * 3

    def update(self, val, count=1):
        self.count += count
        self.sum += val * count
        self.avg = self.sum / self.count

    def __repr__(self):
        text = f"{self.name}: {self.avg:.4f}"
        return text

def get_lr(optimizer):
    for param_group in optimizer.param_groups:
        return param_group["lr"]
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from transformers import DistilBertModel, DistilBertConfig
import timm

# Configuration (CFG) class
class CFG:
    debug = False
    image_path = "/content/drive/MyDrive/ptbxl/generated_ecg_images"
    captions_path = "/content/updated_captions.csv"
    batch_size = 32
    num_workers = 2
    image_encoder_lr = 1e-3
    text_encoder_lr = 1e-3
    head_lr = 5e-4
    weight_decay = 1e-3
    patience = 1
    factor = 0.8
    epochs = 10
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

    model_name = 'resnet50'
    image_embedding = 2048
    text_encoder_model = "distilbert-base-uncased"
    text_embedding = 768
    text_tokenizer = "distilbert-base-uncased"
    max_length = 200
    pretrained = True
    trainable = True
    temperature = 1.0
    size = 224
    projection_dim = 256
    dropout = 0.1

# Define the Image Encoder
class ImageEncoder(nn.Module):
    def __init__(self, model_name=CFG.model_name, pretrained=CFG.pretrained, trainable=CFG.trainable):
        super().__init__()
        self.model = timm.create_model(model_name, pretrained, num_classes=0, global_pool="avg")
        for p in self.model.parameters():
            p.requires_grad = trainable

    def forward(self, x):
        return self.model(x)

# Define the Text Encoder
class TextEncoder(nn.Module):
    def __init__(self, model_name=CFG.text_encoder_model, pretrained=CFG.pretrained, trainable=CFG.trainable):
        super().__init__()
        if pretrained:
            self.model = DistilBertModel.from_pretrained(model_name)
        else:
            self.model = DistilBertModel(config=DistilBertConfig())

        for p in self.model.parameters():
            p.requires_grad = trainable

    def forward(self, input_ids, attention_mask):
        output = self.model(input_ids=input_ids, attention_mask=attention_mask)
        return output.last_hidden_state[:, 0, :]  # CLS token

# Define the Projection Head
class ProjectionHead(nn.Module):
    def __init__(self, embedding_dim, projection_dim=CFG.projection_dim, dropout=CFG.dropout):
        super().__init__()
        self.projection = nn.Linear(embedding_dim, projection_dim)
        self.gelu = nn.GELU()
        self.fc = nn.Linear(projection_dim, projection_dim)
        self.dropout = nn.Dropout(dropout)
        self.layer_norm = nn.LayerNorm(projection_dim)

    def forward(self, x):
        projected = self.projection(x)
        x = self.gelu(projected)
        x = self.fc(x)
        x = self.dropout(x)
        x = x + projected
        x = self.layer_norm(x)
        return x

# Define the CLIP Model
class CLIPModel(nn.Module):
    def __init__(self, temperature=CFG.temperature, image_embedding=CFG.image_embedding, text_embedding=CFG.text_embedding):
        super().__init__()
        self.image_encoder = ImageEncoder()
        self.text_encoder = TextEncoder()
        self.image_projection = ProjectionHead(embedding_dim=image_embedding)
        self.text_projection = ProjectionHead(embedding_dim=text_embedding)
        self.temperature = temperature

    def forward(self, batch):
        # Get Image and Text Features
        image_features = self.image_encoder(batch["image"])
        text_features = self.text_encoder(batch["input_ids"], batch["attention_mask"])

        # Project embeddings
        image_embeddings = self.image_projection(image_features)
        text_embeddings = self.text_projection(text_features)

        # Normalize embeddings for cosine similarity
        image_embeddings = F.normalize(image_embeddings, p=2, dim=-1)
        text_embeddings = F.normalize(text_embeddings, p=2, dim=-1)

        # Compute logits (cosine similarity between image and text embeddings)
        logits = (text_embeddings @ image_embeddings.T) / self.temperature

        # Similarities (cosine) for target construction
        images_similarity = image_embeddings @ image_embeddings.T
        texts_similarity = text_embeddings @ text_embeddings.T

        # Create targets using softmax
        targets = F.softmax((images_similarity + texts_similarity) / 2 * self.temperature, dim=-1).clamp(min=1e-9)

        # Compute the loss for both images and texts
        texts_loss = cross_entropy(logits, targets, reduction='none')
        images_loss = cross_entropy(logits.T, targets.T, reduction='none')

        # Final loss
        loss = (images_loss + texts_loss) / 2.0  # shape: (batch_size)
        return loss.mean()

# Cross entropy function
def cross_entropy(preds, targets, reduction='none'):
    log_softmax = nn.LogSoftmax(dim=-1)
    loss = (-targets * log_softmax(preds)).sum(1)
    if reduction == "none":
        return loss
    elif reduction == "mean":
        return loss.mean()
```

```python
class CLIPDataset(torch.utils.data.Dataset):
    def __init__(self, image_filenames, captions, tokenizer, transforms):
        self.image_filenames = image_filenames
        self.captions = list(captions)
        self.encoded_captions = tokenizer(
            list(captions), padding=True, truncation=True, max_length=CFG.max_length, return_tensors="pt"
        )
        self.transforms = transforms

    def __getitem__(self, idx):
        item = {key: values[idx] for key, values in self.encoded_captions.items()}

        image_path = self.image_filenames[idx]
        if not os.path.exists(image_path):
            print(f"Error: File does not exist at {image_path}")
            item['image'] = torch.zeros((3, CFG.size, CFG.size), dtype=torch.float32)
        else:
            try:
                image = cv2.imread(image_path)
                image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Ensure RGB format
                # Apply the transformation (including ToTensorV2())
                transformed = self.transforms(image=image)
                item['image'] = transformed['image']  # Already converted to tensor
            except Exception as e:
                print(f"Error loading or processing image at index {idx}: {e}")
                item['image'] = torch.zeros((3, CFG.size, CFG.size), dtype=torch.float32)

        item['caption'] = self.captions[idx]
        return item

    def __len__(self):
        return len(self.captions)
```

```python
import pandas as pd
import numpy as np
import torch
from torch.utils.data import DataLoader
from sklearn.model_selection import train_test_split
from transformers import DistilBertTokenizer
from albumentations.pytorch import ToTensorV2
import albumentations as A
from tqdm import tqdm
import itertools

# Function to split the data into training and validation dataframes
def make_train_valid_dfs(dataframe):
    max_id = dataframe["ecg_id"].max() + 1 if not CFG.debug else 100  # Use 'ecg_id' as the identifier

    # Use the full set of image ids
    image_ids = np.arange(0, max_id)

    # Set a random seed for reproducibility
    np.random.seed(42)

    # Split into 80% train and 20% valid sets
    valid_ids = np.random.choice(image_ids, size=int(0.2 * len(image_ids)), replace=False)
    train_ids = [id_ for id_ in image_ids if id_ not in valid_ids]

    # Create train and validation dataframes based on the split ids
    train_dataframe = dataframe[dataframe["ecg_id"].isin(train_ids)].reset_index(drop=True)
    valid_dataframe = dataframe[dataframe["ecg_id"].isin(valid_ids)].reset_index(drop=True)

    return train_dataframe, valid_dataframe


# Training loop for a single epoch
def train_epoch(model, train_loader, optimizer, lr_scheduler, step):
    loss_meter = AvgMeter()
    tqdm_object = tqdm(train_loader, total=len(train_loader))

    for batch in tqdm_object:
        try:
            # Move tensors to device
            batch = {k: v.to(CFG.device) if isinstance(v, torch.Tensor) else v for k, v in batch.items() if k != "caption"}

            loss = model(batch)
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()

            if step == "batch":
                lr_scheduler.step()

            count = batch["image"].size(0)
            loss_meter.update(loss.item(), count)
            tqdm_object.set_postfix(train_loss=loss_meter.avg, lr=get_lr(optimizer))

        except Exception as e:
            print(f"Error processing batch during training: {e}")

    return loss_meter


# Validation loop for a single epoch
def valid_epoch(model, valid_loader):
    loss_meter = AvgMeter()
    tqdm_object = tqdm(valid_loader, total=len(valid_loader))
    for batch in tqdm_object:
        try:
            batch = {k: v.to(CFG.device) if isinstance(v, torch.Tensor) else v for k, v in batch.items() if k != "caption"}
            loss = model(batch)
            count = batch["image"].size(0)
            loss_meter.update(loss.item(), count)
            tqdm_object.set_postfix(valid_loss=loss_meter.avg)
        except Exception as e:
            print(f"Error processing batch during validation: {e}")

    return loss_meter


# Function to define transformations
def get_transforms(mode="train"):
    if mode == "train":
        return A.Compose([
            A.Resize(CFG.size, CFG.size),
            A.HorizontalFlip(p=0.5),
            A.Normalize(max_pixel_value=255.0, always_apply=True),
            ToTensorV2(),
        ])
    else:
        return A.Compose([
            A.Resize(CFG.size, CFG.size),
            A.Normalize(max_pixel_value=255.0, always_apply=True),
            ToTensorV2(),
        ])


# Function to build DataLoaders from the dataframe
def build_loaders(dataframe, tokenizer, mode):
    transforms = get_transforms(mode=mode)

    dataset = CLIPDataset(
        dataframe["filename_lr"].values,
        dataframe["report"].values,
        tokenizer=tokenizer,
        transforms=transforms,
    )

    dataloader = DataLoader(
        dataset,
        batch_size=CFG.batch_size,
        num_workers=CFG.num_workers,
        shuffle=True if mode == "train" else False,
    )

    return dataloader


# Main training function
def main():
    dataframe = captions  # Assuming captions is globally defined or passed
    train_df, valid_df = make_train_valid_dfs(dataframe)
    tokenizer = DistilBertTokenizer.from_pretrained(CFG.text_tokenizer)

    train_loader = build_loaders(train_df, tokenizer, mode="train")
    valid_loader = build_loaders(valid_df, tokenizer, mode="valid")

    model = CLIPModel().to(CFG.device)

    # Setting up optimizer and learning rate scheduler
    params = [
        {"params": model.image_encoder.parameters(), "lr": CFG.image_encoder_lr},
        {"params": model.text_encoder.parameters(), "lr": CFG.text_encoder_lr},
        {"params": itertools.chain(model.image_projection.parameters(), model.text_projection.parameters()), "lr": CFG.head_lr, "weight_decay": CFG.weight_decay}
    ]

    optimizer = torch.optim.AdamW(params, weight_decay=0.)
    '''lr_scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(
        optimizer, mode="min", patience=CFG.patience, factor=CFG.factor
    )'''
    lr_scheduler = torch.optim.lr_scheduler.StepLR(optimizer, step_size=5, gamma=0.5)


    step = "epoch"
    best_loss = float('inf')

    for epoch in range(CFG.epochs):
        print(f"Epoch: {epoch + 1}")
        model.train()
        train_loss = train_epoch(model, train_loader, optimizer, lr_scheduler, step)

        model.eval()
        with torch.no_grad():
            valid_loss = valid_epoch(model, valid_loader)

        if valid_loss.avg < best_loss:
            best_loss = valid_loss.avg
            torch.save(model.state_dict(), "best.pt")
            print("Saved Best Model!")

        lr_scheduler.step(valid_loss.avg)
```

```python
from transformers import CLIPProcessor, CLIPModel
import torch
import numpy as np
from PIL import Image
from tqdm import tqdm  # Import tqdm for progress bar

# Load model directly
from transformers import AutoProcessor, AutoModelForZeroShotImageClassification

processor = AutoProcessor.from_pretrained("openai/clip-vit-base-patch32")
model = AutoModelForZeroShotImageClassification.from_pretrained("openai/clip-vit-base-patch32")


# Step 2: Move the model to the GPU (if available)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)  # Move model to GPU
model.eval()  # Set model to evaluation mode

import pandas as pd
captions = pd.read_csv("updated_captions.csv")
# Assuming you have image paths and text captions (valid_df contains this information)
image_paths = captions["filename_lr"].tolist()
text_captions = captions["report"].tolist()

# Step 3: Define a function to load images
def load_image(image_path):
    return Image.open(image_path)

# Step 4: Batch processing function
def get_embeddings_in_batches(model, processor, image_paths, text_captions, device, batch_size=16):
    all_image_embeddings = []
    all_text_embeddings = []

    # Process images in batches
    print("Generating image embeddings in batches...")
    for i in tqdm(range(0, len(image_paths), batch_size), desc="Images"):
        batch_image_paths = image_paths[i:i + batch_size]
        batch_images = [load_image(image_path) for image_path in batch_image_paths]
        inputs = processor(images=batch_images, return_tensors="pt", padding=True).to(device)  # Move inputs to GPU
        image_embeddings = model.get_image_features(**inputs)
        all_image_embeddings.append(image_embeddings)

    # Process text captions in batches
    print("Generating text embeddings in batches...")
    for i in tqdm(range(0, len(text_captions), batch_size), desc="Text Captions"):
        batch_text_captions = text_captions[i:i + batch_size]
        inputs = processor(text=batch_text_captions, return_tensors="pt", padding=True).to(device)  # Move inputs to GPU
        text_embeddings = model.get_text_features(**inputs)
        all_text_embeddings.append(text_embeddings)

    # Convert list of embeddings to tensors
    all_image_embeddings = torch.cat(all_image_embeddings).to(device)
    all_text_embeddings = torch.cat(all_text_embeddings).to(device)

    return all_image_embeddings, all_text_embeddings

# Step 5: Calculate similarity and MRR with progress bar
def compute_mrr_at_k_clip(all_image_embeddings, all_text_embeddings, text_captions, K_list=[10, 50, 100, 500]):
    mrr_at_k = {K: [] for K in K_list}

    # Normalize embeddings
    all_image_embeddings = torch.nn.functional.normalize(all_image_embeddings, p=2, dim=-1)
    all_text_embeddings = torch.nn.functional.normalize(all_text_embeddings, p=2, dim=-1)

    # Compute similarity between all image-text pairs
    similarity_matrix = torch.matmul(all_image_embeddings, all_text_embeddings.T)  # [N, N] matrix

    # For each image, find the rank of the correct caption
    print("Calculating MRR...")
    for i in tqdm(range(len(text_captions)), desc="MRR Calculation"):
        correct_caption = text_captions[i]

        # Get similarity scores for the current image
        scores = similarity_matrix[i].cpu().numpy()

        # Sort indices based on similarity scores (higher score = more similar)
        sorted_indices = np.argsort(scores)[::-1]  # Sort descending

        # Find rank of the correct text
        correct_rank = np.where(sorted_indices == i)[0][0] + 1  # Find correct rank (+1 to convert 0-index to 1-index)

        # Calculate MRR@K for each K
        for K in K_list:
            if correct_rank <= K:
                mrr_at_k[K].append(1 / correct_rank)
            else:
                mrr_at_k[K].append(0)

    # Compute the mean MRR for each K
    for K in K_list:
        mrr_at_k[K] = np.mean(mrr_at_k[K])
        print(f"MRR@{K}: {mrr_at_k[K]:.4f}")

    return mrr_at_k

# Step 6: Generate embeddings with batch processing
batch_size = 16  # Adjust this based on available GPU memory
all_image_embeddings, all_text_embeddings = get_embeddings_in_batches(model, processor, image_paths, text_captions, device, batch_size)

# Step 7: Compute MRR for CLIP with progress bar
mrr_at_k_clip = compute_mrr_at_k_clip(all_image_embeddings, all_text_embeddings, text_captions, K_list=[10, 50, 100, 500])
```
